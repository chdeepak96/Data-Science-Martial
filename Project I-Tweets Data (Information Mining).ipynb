```python
#Focus on Text field
#Preprocess data
#Remove PUNCTUATION,STOPWORDS,LOWER CASE

import string
punctuation=string.punctuation
```

```python
from nltk.corpus import stopwords
stopwords=stopwords.words("english")
```

```python
def _cleaned(text):
    text=text.lower()
    text= "".join(w for w in text if w not in punctuation)
    word=text.split()
    text=" ".join(i for i in word if i not in stopwords)
    
    return text
```

```python
#To find top named entities in text
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
```

```python
for text in tweets["text"]:
    entities = ne_chunk(pos_tag(word_tokenize(text)))
    for entity in entities:
        if hasattr(entity,"label"):
            print(entity)
```
